```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
Obj = ['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope']
for i in Obj:
    hrt[i] = le.fit_transform (hrt[i])
hrt.head()
```

```python
x = hrt.drop(['Sex', 'ChestPainType', 'FastingBS', 'RestingECG', 'ExerciseAngina', 'Oldpeak', 'ST_Slope', 'HeartDisease'], axis =1)
x
```

```python
y = hrt[['Sex', 'ChestPainType', 'FastingBS', 'RestingECG', 'ExerciseAngina', 'Oldpeak', 'ST_Slope', 'HeartDisease']]
y
```

```python
from sklearn.preprocessing import MinMaxScaler
MMS = MinMaxScaler()
x1 = MMS.fit_transform(x)
x1
```

```python
x1.shape
```

```python
x1 = pd.DataFrame(x1, columns=['Age', 'RestingBP', 'Cholesterol', 'MaxHR'])
x1
```

```python
x1.describe()
```

```python
x2 = pd.concat([x1, y], axis=1)
x2
```

```python
# Logistic Regression Model:

x = x2.drop('HeartDisease', axis = 1)
y = x2 ['HeartDisease']
x_1 = pd.DataFrame(x)
y_1 = pd.DataFrame(y)
print(y_1)
print(x_1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.25,random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
Log_r = LogisticRegression(max_iter = 200)
```

```python
Model = Log_r.fit(x_train, y_train)
```

```python
y_pred = Model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, accuracy_score, f1_score, classification_report
```

```python
print (confusion_matrix(y_test, y_pred))
```

```python
print (precision_score(y_test,y_pred))
```

```python
print (recall_score(y_test, y_pred))
```

```python
# 'should do'
print (accuracy_score(y_test,y_pred))
```

```python
print(f1_score(y_test,y_pred))
```

```python
print (classification_report(y_test,y_pred))
```
