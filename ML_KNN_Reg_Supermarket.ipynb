```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x = ['Invoice ID', 'Branch', 'City', 'Customer type', 'Gender', 'Product line', 'Date', 'Time', 'Payment']
for i in x:
    sup[i] = le.fit_transform (sup[i])
sup.head()
```

```python
x = sup.drop(['Invoice ID', 'Branch', 'City', 'Customer type', 'Gender', 'Product line', 'Date', 'Time', 'Payment'], axis =1)
x
```

```python
y = sup [['Invoice ID', 'Branch', 'City', 'Customer type', 'Gender', 'Product line', 'Date', 'Time', 'Payment']]
y
```

```python
from sklearn.preprocessing import MinMaxScaler
MMS = MinMaxScaler()
x1 = MMS.fit_transform(x)
x1
```

```python
y1 = pd.DataFrame (x1, columns=['Unit price', 'Quantity', 'Tax 5%', 'Total', 'cogs', 
                                'gross margin percentage', 'gross income', 'Rating'])
y1
```

```python
y1.describe()
```

```python
x2 = pd.concat ([y, y1], axis = 1)
x2
```

```python
a = x2.drop ('Unit price', axis = 1)
a1 = pd.DataFrame (a)
print (a1)
b = x2 ['Unit price']
b1 = pd.DataFrame (b)
print (b1)
```

```python
from sklearn.model_selection import train_test_split
a1_train, a1_test, b1_train, b1_test = train_test_split (a1, b1, test_size = 0.25, random_state =42)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score
k_metric = []
neighbors = np.arange (1, 15)
for k in neighbors:
    KNN = KNeighborsRegressor(n_neighbors = k)
    model = KNN.fit (a1_train, b1_train)
    pred_KNN = model.predict (a1_test)
    score = r2_score (b1_test, pred_KNN)
    k_metric.append(score)
```

```python
k_metric
```

```python
import matplotlib.pyplot as plt
plt.plot (neighbors, k_metric,'o-')
plt.xlabel ('K Value')
plt.ylabel ('Accuracy')
```

```python
KNN = KNeighborsRegressor (n_neighbors = 6)
model = KNN.fit (a1_train, b1_train)
pred_KNN = model.predict (a1_test)
```

```python
from sklearn.metrics import mean_squared_error
print ('mse:', mean_squared_error(b1_test, pred_KNN))
from sklearn.metrics import r2_score
print ('r2_score:', r2_score (b1_test, pred_KNN))
```
