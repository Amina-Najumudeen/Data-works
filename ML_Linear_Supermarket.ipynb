```python
# Step 1: Null Handling 

spt.isnull().sum()
```

```python
# Step 2: Encoding

from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
for i in ['Invoice ID', 'Branch', 'City', 'Customer type', 'Gender', 'Product line', 'Date', 'Time', 'Payment']:
    spt[i] = le.fit_transform (spt[i])
```

```python
spt.head()
```

```python
# Step 3: Scaling

from sklearn.preprocessing import MinMaxScaler
```

```python
MMS = MinMaxScaler()
```

```python
out = spt.drop (['Total', 'Invoice ID', 'Branch', 'City', 'Customer type', 'Gender', 'Product line', 
                 'Date', 'Time', 'Payment'], axis = 1)
out
```

```python
inp = spt [['Total', 'Invoice ID', 'Branch', 'City', 'Customer type', 'Gender', 'Product line', 
                 'Date', 'Time', 'Payment']]
inp
```

```python
out1 = MMS.fit_transform (out)
out1
```

```python
type(out1)
```

```python
spt.dtypes
```

```python
out1 = pd.DataFrame (out1, columns = ['Unit price', 'Quantity', 'Tax 5%', 'cogs', 'gross margin percentage', 'gross income', 'Rating'])
out1
```

```python
out1.describe()
```

```python
out2 = pd.concat ([out1, inp], axis =1)
out2
```

```python
out2.corr()
```

```python
x = out2.drop('Total',axis=1)
x1 = pd.DataFrame(x)
print(x1)
y = out2 ['Total']
y1 = pd.DataFrame(y)
print(y1)
```

```python
from sklearn.model_selection import train_test_split
x1_train, x1_test, y1_train, y1_test = train_test_split (x1, y1, test_size = 0.35, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
model = lr.fit(x1_train, y1_train)
```

```python
y1_pred = model.predict (x1_test)
print (y1_pred)
```

```python
print (y1_test)
```

```python
from sklearn.metrics import mean_squared_error
print ('MSE is :' , mean_squared_error (y1_test, y1_pred)) 
```

```python
from sklearn.metrics import r2_score
print ('r2 score:' , r2_score (y1_test, y1_pred))
```
